```python
import os
from dotenv import load_dotenv
import pandas as pd
from lyricsgenius import Genius
import time

def get_lyrics_by_rapper(rapper_name, max_songs=5):
    lyrics_data = []
    
    try:
        artist = genius.search_artist(rapper_name, max_songs=max_songs, sort="popularity")
        if artist is None:
            print(f"Artist '{rapper_name}' not found.")
            return lyrics_data

        for song in artist.songs:
            song_data = song._body  # Raw Genius JSON data
            release_date = song_data.get("release_date", "N/A")

            lyrics_data.append({
                'track_name': song.full_title,
                'artist': song.artist,
                'lyrics': song.lyrics,
                'release_date': release_date
            })
    except Exception as e:
        print(f"Error fetching data for {rapper_name}: {e}")

    return lyrics_data

if __name__ == "__main__":

    # Load env variables
    load_dotenv()
    GENIUS_API_TOKEN = os.getenv("GENIUS_API_TOKEN")
    
    # Create genius object
    genius = Genius(GENIUS_API_TOKEN, skip_non_songs=True, excluded_terms=["(Remix)", "(Live)"], remove_section_headers=True, retries=5)

    # Get rapper names from csv
    rapper_names = pd.read_csv('../data/rapper_names.csv', index_col=0)

    all_lyrics = []
    # Iterate through rapper names and get all lyrics from their songs
    for rapper in rapper_names['Name'].unique()[:10]:
        print(f'Getting lyrics for {rapper}')

        # 
        data = get_lyrics_by_rapper(rapper)

        # If rapper not found
        if not data:
            print(f"No data found for {rapper}")
            continue

        all_lyrics.extend(data)
        time.sleep(3)
    
    # Save as one CSV
    df = pd.DataFrame(all_lyrics)
    df.to_csv("../data/all_lyrics.csv", index=False, encoding='utf-8')
        


```
